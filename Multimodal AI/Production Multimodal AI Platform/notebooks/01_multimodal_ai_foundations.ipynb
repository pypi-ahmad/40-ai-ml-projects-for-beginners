# Multimodal AI Foundations and Production Build

This notebook explains core multimodal concepts before implementation.

## 1. What is Multimodal AI?
Multimodal AI combines text, image, document, and layout signals for joint reasoning.

## 2. Vision-Language Models
We support CLIP, SigLIP, Florence-2, Qwen2.5-VL, Llama Vision, MiniCPM-V, BLIP-2, SmolVLM.

## 3. OCR + Document Intelligence
Pipeline uses native parsing first, then OCR fallback for scanned documents.

## 4. Semantic Embeddings and Retrieval
CLIP/SigLIP embeddings indexed in ChromaDB enable semantic image and document search.

## 5. Multimodal RAG
Upload -> OCR -> chunk -> embed -> retrieve -> answer with grounded context.

```python
from multimodal_ai.services.bootstrap import build_platform_service
from multimodal_ai.domain import RequestEnvelope, InputPayload, TraceContext

service = build_platform_service()
service.health().model_dump()
```

## 6. Captioning Example

```python
request = RequestEnvelope(
    input=InputPayload(image_path='data/uploads/example.png'),
    options={'style': 'detailed'},
    trace=TraceContext(source='notebook'),
)
service.caption(request).model_dump()
```

## 7. Visual Question Answering Example

```python
request = RequestEnvelope(
    input=InputPayload(image_path='data/uploads/example.png', question='What objects are visible?'),
    trace=TraceContext(source='notebook'),
)
service.vqa(request).model_dump()
```

## 8. OCR and RAG Example

```python
request = RequestEnvelope(
    input=InputPayload(document_path='data/uploads/example.pdf'),
    trace=TraceContext(source='notebook'),
)
service.documents(request).model_dump()
```
