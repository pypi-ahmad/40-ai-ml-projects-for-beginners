# Zero-to-Hero: Production Internet AI Agent

This notebook teaches the architecture and implementation of a production-grade internet-connected agent with LangGraph, Ollama, tools, retrieval, memory, verification, and deployment surfaces.

## 1) Environment and Imports

Load core modules used across API, CLI, and Streamlit.

```python
from internet_agent.config import get_settings
from internet_agent.services.agent_service import InternetAgentService

settings = get_settings()
settings
```

## 2) LangGraph Multi-Agent Workflow

The workflow chains 10 agents with conditional verification retries.

```python
service = InternetAgentService(settings=settings)
service.workflow.graph
```

## 3) Search Planning and Retrieval

Run hybrid retrieval: provider search -> extraction -> chunking -> embeddings -> ChromaDB.

```python
import asyncio

payload = asyncio.run(service.search(session_id='nb', query='latest Python version'))
len(payload['results']), len(payload['documents']), len(payload['chunks'])
```

## 4) End-to-End Chat with Verification

Run the full multi-agent graph including verification and reflection.

```python
result = asyncio.run(service.chat(session_id='nb', message='Compare today\'s NVIDIA news with yesterday.'))
result['confidence'], result['hallucination_risk']
```

## 5) Memory and Semantic Retrieval

Inspect persisted history and query semantic memory.

```python
history = service.history('nb')
hits = service.memory_search('nvidia news', top_k=5)
len(history['messages']), len(hits['hits'])
```

## 6) Report Generation

Export Markdown/HTML/PDF/JSON reports with citations and reasoning traces.

```python
report = service.export_report('nb', result['report'], fmt='json')
report
```
