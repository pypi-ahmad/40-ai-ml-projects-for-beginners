# 03 — Exploratory Data Analysis (EDA)

## Definition
EDA is structured investigation of data distribution, quality, seasonality, and relationships before modeling.

## Theory
For forecasting, EDA must include temporal decomposition and behavior patterns.

## Motivation
Good EDA drives better features and fewer modeling surprises.

## Common Mistakes
- ignoring temporal order
- trusting aggregate metrics only
- skipping distribution tails and outliers

```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from modules.data_generator import run_data_augmentation
from modules.settings import load_config

cfg = load_config()
aug = run_data_augmentation(cfg)
aug['Date'] = pd.to_datetime(aug['Date'])
print('Augmented shape:', aug.shape)
aug.head()
```

```
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

sns.histplot(aug['Usage (minutes)'], kde=True, ax=axes[0,0])
axes[0,0].set_title('Usage Distribution')

sns.boxplot(data=aug, x='App', y='Usage (minutes)', ax=axes[0,1])
axes[0,1].tick_params(axis='x', rotation=45)
axes[0,1].set_title('Usage by App')

daily = aug.groupby('Date')['Usage (minutes)'].mean().reset_index()
axes[1,0].plot(daily['Date'], daily['Usage (minutes)'])
axes[1,0].set_title('Daily Average Usage Trend')

aug['day_of_week'] = aug['Date'].dt.day_name()
order = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']
sns.boxplot(data=aug, x='day_of_week', y='Usage (minutes)', order=order, ax=axes[1,1])
axes[1,1].tick_params(axis='x', rotation=30)
axes[1,1].set_title('Weekly Pattern')

plt.tight_layout()
plt.show()
```

```
corr_cols = ['Usage (minutes)', 'Notifications', 'Times Opened']
print('Correlation matrix:')
print(aug[corr_cols].corr().round(3))

agg = aug.groupby('App')[['Usage (minutes)','Notifications','Times Opened']].mean().round(2)
agg.sort_values('Usage (minutes)', ascending=False)
```

## Insights and Interpretation
- Entertainment and messaging apps show higher average usage than professional apps.
- Weekly seasonality appears with weekend lift.
- Notifications and opens correlate with usage, motivating behavioral features.
