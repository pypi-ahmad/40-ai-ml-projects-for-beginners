# 04 — Feature Engineering and Feature Selection

## Definition
Feature engineering converts raw events into predictive signals.

## Theory
Time-aware systems require lag/rolling/calendar context to predict future behavior.

## Motivation
Without temporal and behavioral context, regression models underfit engagement dynamics.

## Feature Families
- Temporal: day/week/month/quarter/weekend/holiday
- Lag: prior usage/notifications/opens
- Rolling: moving averages and volatility
- Behavior: notifications per usage minute, open frequency, engagement score
- Interaction: notifications×opens, usage×opens

```
from modules.data_generator import run_data_augmentation
from modules.feature_engineering import build_feature_pipeline
from modules.settings import load_config

cfg = load_config()
aug = run_data_augmentation(cfg)
feat = build_feature_pipeline(aug, cfg)
print('Feature shape:', feat.shape)
feat.head()
```

```
from modules.feature_selector import run_feature_selection

ranking = run_feature_selection(feat, config=cfg)
ranking.head(20)
```

```
import matplotlib.pyplot as plt

top = ranking.head(15).sort_values('composite_score')
plt.figure(figsize=(8, 6))
plt.barh(top['feature'], top['composite_score'])
plt.title('Top Composite Feature Scores')
plt.xlabel('Composite Score')
plt.tight_layout()
plt.show()
```

## Best Practices
- compute lag/rolling inside entity groups only
- shift target to future horizon before modeling
- log selection outputs for auditability

## Common Mistakes
- leakage via global target encoding
- random split before feature generation
- removing features using one method only
