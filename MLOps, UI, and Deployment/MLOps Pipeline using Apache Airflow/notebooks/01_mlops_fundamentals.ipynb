# 01 — MLOps Fundamentals (Zero-to-Hero)

## Definition
MLOps is the discipline of building reliable ML systems by combining data engineering, model engineering, and operations.

## Theory
A production ML system has five connected lifecycles: Data Pipeline, Feature Pipeline, Model Pipeline, Deployment Pipeline, and Monitoring Pipeline.

## Motivation
Notebook-only ML often fails in production because data shifts, models decay, and manual retraining does not scale.

## Real-World Example
A mobile app engagement predictor must retrain on new behavior every week and raise alerts when usage patterns drift.

## Visual Explanation
```text
Raw Data -> Validation -> Features -> Training -> Registry -> Monitoring -> Retraining
```

## Best Practices
- version data, code, config, and model artifacts
- automate recurring tasks
- monitor model quality and drift continuously

## Common Mistakes
- random train/test split for time-series data
- fitting preprocessing on full dataset (leakage)
- no model lineage or experiment history


## Traditional ML vs MLOps

| Aspect | Traditional ML | MLOps |
|---|---|---|
| Training | Manual | Scheduled/automated |
| Reproducibility | weak | strong |
| Monitoring | ad-hoc | built-in |
| Deployment readiness | low | high |
| Incident response | reactive | proactive |

````
import pandas as pd
from modules.data_loader import load_dataset
from modules.settings import load_config

cfg = load_config()
raw = load_dataset('raw', config=cfg)
print('Raw shape:', raw.shape)
print('Columns:', raw.columns.tolist())
raw.head()
````

````
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(14, 4))
raw['Usage (minutes)'].hist(ax=axes[0], bins=25)
axes[0].set_title('Usage Distribution')
raw.groupby('App')['Usage (minutes)'].mean().sort_values().plot(kind='bar', ax=axes[1])
axes[1].set_title('Average Usage by App')
axes[1].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()
````
