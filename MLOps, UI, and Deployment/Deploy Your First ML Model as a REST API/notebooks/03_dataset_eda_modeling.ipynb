```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

from src.data import load_california_housing, split_dataset
from src.evaluation import compute_regression_metrics

X, y = load_california_housing()
print('Shape:', X.shape)
display(X.head())
display(y.head())
split = split_dataset(X, y, random_state=42)
print('Train/Val/Test:', len(split.X_train), len(split.X_val), len(split.X_test))
```

## Metric Theory

- MAE: average absolute error, robust interpretation.
- MSE: squared error, stronger penalty for large mistakes.
- RMSE: MSE root, same unit as target.
- R²: explained variance proportion.
- MAPE: percentage error, scale independent.

```
model = LinearRegression()
model.fit(split.X_train, split.y_train)
pred = model.predict(split.X_val)
metrics = compute_regression_metrics(split.y_val, pred)
print(metrics)
```
