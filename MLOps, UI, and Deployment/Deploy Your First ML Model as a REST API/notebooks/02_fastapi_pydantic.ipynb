
## FastAPI Fundamentals

### Definition
FastAPI Fundamentals in this project means we design, validate, serve, and observe ML predictions through stable HTTP interfaces.

### Theory
APIs decouple producers (model service) and consumers (web app, batch jobs, partner systems). Typed contracts reduce ambiguity and production failures.

### Motivation
FastAPI gives type-driven validation and auto docs, reducing boilerplate in ML service development.

### Real-World Example
Internal recommendation microservices often standardize on FastAPI for speed + OpenAPI compatibility.

### Best Practices
- Keep contracts explicit and versioned.
- Validate early at boundary.
- Log request IDs for traceability.
- Measure latency, error rate, and throughput.

### Common Mistakes
- Skipping response models and returning arbitrary dicts.
- Pushing business logic directly into route handlers.


## Framework Tradeoffs

- Flask: flexible, minimal, but manual typing/docs.
- FastAPI: strong typing, async-ready, automatic docs.
- Django: batteries included; heavier for pure inference APIs.
- Node.js APIs: high ecosystem breadth; type safety depends on stack choice.

```
from pydantic import BaseModel, Field, ValidationError

class HousingRecord(BaseModel):
    MedInc: float = Field(ge=0.0, le=30.0)
    HouseAge: float = Field(ge=0.0, le=100.0)
    AveRooms: float = Field(ge=0.1, le=100.0)
    AveBedrms: float = Field(ge=0.1, le=20.0)
    Population: float = Field(ge=1.0, le=100000.0)
    AveOccup: float = Field(ge=0.1, le=100.0)
    Latitude: float = Field(ge=32.0, le=43.0)
    Longitude: float = Field(ge=-125.0, le=-113.0)

valid = {
    'MedInc': 8.3252, 'HouseAge': 41.0, 'AveRooms': 6.9841,
    'AveBedrms': 1.0238, 'Population': 322.0, 'AveOccup': 2.5556,
    'Latitude': 37.88, 'Longitude': -122.23
}
print('Valid model:', HousingRecord(**valid).model_dump())

invalid = dict(valid)
invalid['Longitude'] = -200.0
try:
    HousingRecord(**invalid)
except ValidationError as exc:
    print('Validation error snippet:')
    print(exc.errors()[0])
```
