## 4) Local Models vs Cloud APIs

**Local model strengths**:
- data privacy
- no per-token billing
- offline operation

**Tradeoff**:
- hardware constraints
- local deployment/ops complexity


```python
import pandas as pd

comparison = pd.DataFrame(
    [
        ('Privacy', 'High (data stays local)', 'Depends on provider policy'),
        ('Cost model', 'Fixed hardware cost', 'Usage-based API billing'),
        ('Latency consistency', 'Hardware dependent', 'Network + provider dependent'),
        ('Operational complexity', 'Higher local ops', 'Lower local ops'),
    ],
    columns=['Dimension', 'Local inference', 'Cloud API'],
)
comparison
```
