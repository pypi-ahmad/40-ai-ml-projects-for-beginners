## 1) Definition, Theory, Motivation, and Real-world Framing

**Definition**: AI application = software where model inference is a first-class runtime dependency.

**Theory**: AI output is probabilistic. Reliability comes from architecture + validation + guardrails,
not only model size.

**Motivation**: You need UX, prompt design, caching, and telemetry to ship AI features safely.

**Real-world examples**:
- support ticket triage
- sales note summarization
- multilingual customer communication
- document extraction and compliance checks

**Best practices**:
- keep data/model/inference boundaries explicit
- always validate and parse outputs
- add fallbacks and error messages for model downtime

**Common mistakes**:
- directly rendering raw model output to users
- no timeout/retry policy
- no benchmark evidence before model choice

## 2) Visual Explanation — Application Layers

The following code draws a simple architecture figure and saves it to `outputs/figures`.


```python
from pathlib import Path
import matplotlib.pyplot as plt

fig_dir = Path('outputs/figures')
fig_dir.mkdir(parents=True, exist_ok=True)

labels = [
    'Frontend\n(Streamlit)',
    'Orchestration\n(Validation + State)',
    'Model Services\n(Prompts + Parsing)',
    'Inference\n(Ollama Local Models)',
    'Artifacts\n(Metrics + Figures)',
]

fig, ax = plt.subplots(figsize=(8, 6))
ax.axis('off')

y = 0.9
for i, label in enumerate(labels):
    ax.text(0.5, y, label, ha='center', va='center', fontsize=11,
            bbox={'boxstyle': 'round,pad=0.4', 'facecolor': '#e0f2fe', 'edgecolor': '#0284c7'})
    if i < len(labels) - 1:
        ax.annotate('', xy=(0.5, y - 0.08), xytext=(0.5, y - 0.02),
                    arrowprops={'arrowstyle': '->', 'lw': 1.5, 'color': '#334155'})
    y -= 0.16

output_path = fig_dir / '00_ai_app_layers.png'
fig.tight_layout()
fig.savefig(output_path, dpi=180)
plt.close(fig)
output_path
```

## 3) User Interaction Flow

**Flow**:
1. User enters input
2. Validation layer checks completeness
3. Prompt builder formats instruction
4. Model returns response
5. Parser normalizes output
6. UI renders result and metrics

**Common failure modes**:
- empty input
- malformed JSON from model
- inference timeout
- unsupported file format


```python
from dataclasses import dataclass

@dataclass
class AppFlowEvent:
    step: str
    status: str

flow = [
    AppFlowEvent('Input capture', 'ok'),
    AppFlowEvent('Validation', 'ok'),
    AppFlowEvent('Prompt build', 'ok'),
    AppFlowEvent('Inference call', 'pending if model unavailable'),
    AppFlowEvent('Output parsing', 'ok with fallback'),
    AppFlowEvent('Render', 'ok'),
]

[event.__dict__ for event in flow]
```
