## 1) Streamlit Architecture and Rerun Model

**Definition**: Streamlit executes your script top-to-bottom whenever widget state changes.

**Theory**: Session persistence requires `st.session_state`.

**Motivation**: AI app UIs feel broken if chat history and settings reset on every click.

**Best practices**:
- initialize all session keys once
- keep expensive resources cached
- keep page modules focused and small

**Common mistakes**:
- doing expensive model calls on every rerun
- forgetting widget keys in repeated components

## 3) Caching Fundamentals (`st.cache_resource` vs `st.cache_data`)

**Definition**:
- `st.cache_resource`: cache singleton-like resources (clients, model handles)
- `st.cache_data`: cache deterministic function outputs

**Theory**: caching converts repeated expensive operations into lookups.

**Common mistakes**: caching mutable objects with hidden side-effects.


```python
import time
from functools import lru_cache

def slow_square(x: int) -> int:
    time.sleep(0.1)
    return x * x

@lru_cache(maxsize=None)
def cached_square(x: int) -> int:
    time.sleep(0.1)
    return x * x

t0 = time.perf_counter(); slow_square(10); t1 = time.perf_counter()
t2 = time.perf_counter(); cached_square(10); t3 = time.perf_counter()
t4 = time.perf_counter(); cached_square(10); t5 = time.perf_counter()

{
    'uncached_seconds': round(t1 - t0, 4),
    'cached_first_seconds': round(t3 - t2, 4),
    'cached_second_seconds': round(t5 - t4, 6),
}
```
