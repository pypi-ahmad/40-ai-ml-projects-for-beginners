## 1) Testing Principles for AI Applications

**Definition**: testing AI apps combines deterministic software tests and probabilistic behavior checks.

**Theory**: model output can vary, so test parsing, contracts, and failure handling deterministically.

**Best practices**:
- mock model calls for unit tests
- keep optional live integration tests separate
- validate input and schema boundaries

**Common mistakes**:
- testing only happy path
- asserting exact free-form model text


````python
import subprocess

cmd = ['bash', '-lc', 'UV_CACHE_DIR=.uv_cache uv run pytest -q']
proc = subprocess.run(cmd, capture_output=True, text=True)

{
    'returncode': proc.returncode,
    'stdout_tail': '\n'.join(proc.stdout.strip().splitlines()[-5:]),
    'stderr_tail': '\n'.join(proc.stderr.strip().splitlines()[-5:]),
}
````
