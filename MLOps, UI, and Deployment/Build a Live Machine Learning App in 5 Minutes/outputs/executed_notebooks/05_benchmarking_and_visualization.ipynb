## Why Benchmarking Matters

### Definition
Benchmarking quantifies tradeoffs between speed, quality, and resource cost.

### Theory
Model selection should be evidence-driven, not preference-driven.

### Motivation
Without benchmarks, scaling decisions become guesswork.

### Real-World Examples
Choosing chat model based only on one sample often fails in production.

### Visual Explanation
```
Measure -> Compare -> Decide -> Document
```

### Code Explanation
`BenchmarkRunner` produces structured metrics and exports JSON/CSV artifacts.

### Best Practices
Benchmark multiple prompt sizes to surface scaling behavior.

### Common Mistakes
Using single run and assuming stable latency.


````python
from src.benchmarking import BenchmarkRunner, format_benchmark_table

runner = BenchmarkRunner()
medium_results = runner.run_all(prompt_key='medium', runs=3)
short_results = runner.run_all(prompt_key='short', runs=3)
long_results = runner.run_all(prompt_key='long', runs=3)
artifacts = runner.export_results(medium_results)
runner.close()

print(format_benchmark_table(medium_results))
artifacts
````

## Visualization for Decision-Making

### Definition
Charts convert raw metrics into actionable model-selection insights.

### Theory
Different chart types answer different questions: bars for comparisons, grouped bars for scaling, radar for multidimensional tradeoffs.

### Motivation
Interviewers and stakeholders can quickly interpret model behavior through visuals.

### Real-World Examples
Latency chart highlights fastest model, radar chart highlights balanced model.

### Visual Explanation
```
metrics -> plotting -> artifact files -> README report
```

### Code Explanation
`BenchmarkVisualizer` generates all figures into `outputs/figures/`.

### Best Practices
Keep consistent axes and units for fair comparisons.

### Common Mistakes
Comparing metrics with inconsistent scales or hidden units.


````python
from src.visualization import BenchmarkVisualizer

visualizer = BenchmarkVisualizer(output_dir='outputs/figures')
figure_paths = visualizer.generate_all(
    medium_results=medium_results,
    short_results=short_results,
    long_results=long_results,
)
figure_paths
````
