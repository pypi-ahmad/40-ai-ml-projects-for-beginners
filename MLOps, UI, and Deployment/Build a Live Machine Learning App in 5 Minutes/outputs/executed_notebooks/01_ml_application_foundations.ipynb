## Traditional Software vs ML Applications

### Definition
Traditional software uses deterministic logic. ML applications use statistical models that infer outputs from learned patterns.

### Theory
In traditional systems, output is fixed for same input. In ML systems, output quality depends on training data, model architecture, and runtime inference settings.

### Motivation
Without this distinction, teams often debug ML bugs as if they were regular code bugs and miss data/model root causes.

### Real-World Examples
Spam filtering, OCR, recommendation systems, and AI chat assistants are all ML applications with uncertainty-aware outputs.

### Visual Explanation
```
User Input -> Business Logic -> Fixed Output   (traditional)
User Input -> Preprocess -> Model Inference -> Probabilistic Output   (ML)
```

### Code Explanation
Later notebooks map each stage into concrete Python modules and Gradio callbacks.

### Best Practices
Track confidence, latency, and fallback paths in addition to correctness.

### Common Mistakes
Assuming model output is always deterministic and never validating edge inputs.


## Five-Layer ML Application Architecture

### Definition
A production ML app has UI, model, inference, data, and deployment layers.

### Theory
Separation of concerns keeps experimentation fast while preserving reliability.

### Motivation
Clear layering enables independent scaling: UI updates should not require retraining models.

### Real-World Examples
Customer support copilots: UI for agents, inference layer for prompts, model layer for LLM, data layer for documents, deployment layer for serving.

### Visual Explanation
```
[UI Layer] -> [Inference Layer] -> [Model Layer]
      |               |               |
      +-------> [Data Layer] <--------+
                     |
             [Deployment Layer]
```

### Code Explanation
`app.py` maps UI interactions. `src/*.py` modules implement model + inference logic. Deployment settings define runtime access.

### Best Practices
Keep contracts stable between layers using typed schemas.

### Common Mistakes
Mixing prompt construction and UI rendering in one function.

## Request-Response Flow in This Project

### Definition
User input passes through validation, preprocessing, inference, postprocessing, then UI rendering.

### Theory
Each stage should return structured outputs and explicit errors to avoid silent failures.

### Motivation
This flow is reusable across sentiment, summarization, translation, chat, and document analysis.

### Real-World Examples
If a user uploads PDF: extract text -> summarize -> optional Q&A -> show outputs in tab.

### Visual Explanation
```
Input -> Validate -> Build Prompt -> Ollama API -> Parse -> UI
```

### Code Explanation
All task modules follow this contract with pydantic models in `src/schemas.py`.

### Best Practices
Capture latency and model metadata in every response.

### Common Mistakes
Returning raw exceptions directly to users without readable context.
