## From Architecture to App Implementation

### Definition
Implementation maps each task into dedicated module and one tab callback.

### Theory
Module-per-task design improves maintainability and testability.

### Motivation
Product-grade app needs extensibility for adding new tabs/features.

### Real-World Examples
Adding speech-to-text later should not require rewriting sentiment/chat modules.

### Visual Explanation
```
Tab -> Handler -> Task Module -> Ollama Client -> Typed Result
```

### Code Explanation
`app.py` wires tabs and keeps handlers lightweight; heavy logic lives in `src/`.

### Best Practices
Keep handler contracts explicit: input types, output shapes, error payload.

### Common Mistakes
Embedding core business logic directly inside UI callback bodies.

## State Management for Chat and Settings

### Definition
State stores mutable session data such as conversation history and selected controls.

### Theory
State must be bounded to prevent memory leaks.

### Motivation
Reliable state handling is key for multi-turn user trust.

### Real-World Examples
Chat tab preserves context while enforcing max-turn trimming.

### Visual Explanation
```
history_state -> callback -> updated_history_state
```

### Code Explanation
`ChatEngine` loads serialized state, trims history, returns updated turns.

### Best Practices
Test state transitions with deterministic stubs.

### Common Mistakes
Letting unbounded chat history grow forever.


## UX Design Patterns for AI Interfaces

### Definition
Good AI UX combines clarity, responsiveness, and graceful failure messaging.

### Theory
Users must understand what model does, expected input, and how to recover from errors.

### Motivation
Strong UX differentiates portfolio project from toy demo.

### Real-World Examples
Progress indicators, placeholders, and contextual helper text reduce friction.

### Visual Explanation
```
Clear input -> Explain action -> Show result + metadata
```

### Code Explanation
App uses structured markdown errors and exposes debug JSON for advanced users.

### Best Practices
Provide defaults aligned to common user paths.

### Common Mistakes
Returning cryptic stack traces in UI output.
