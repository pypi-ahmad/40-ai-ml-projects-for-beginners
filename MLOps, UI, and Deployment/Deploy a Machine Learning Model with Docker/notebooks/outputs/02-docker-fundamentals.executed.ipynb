## 2.2 VM vs Container
**Definition:** VM virtualizes hardware with full guest OS; container virtualizes userspace on shared kernel.

**Theory:** Explain mechanism, assumptions, and where this fits in deployment lifecycle.

**Motivation:** Why this matters for reliability, reproducibility, and operations.

**Real-World Example:** API startup in seconds with container vs slower full VM boot.

**Visual Explanation:** See figure/code cell below.

**Code Explanation:** Code cell demonstrates concrete implementation details.

**Best Practices:** Use containers for app packaging and VMs for stronger isolation boundaries.

**Common Mistakes:** Assuming containers provide same isolation guarantees as full VMs.

```python
docker_commands = {
    'build': 'docker build -f docker/Dockerfile.optimized -t ml-api:latest .',
    'run': 'docker run -p 8001:8000 ml-api:latest',
    'logs': 'docker logs <container-id>',
    'exec': 'docker exec -it <container-id> sh',
    'inspect': 'docker inspect <container-id>',
    'stop': 'docker stop <container-id>',
    'rm': 'docker rm <container-id>'
}
for name, cmd in docker_commands.items():
    print(f'{name:>8}: {cmd}')
```
