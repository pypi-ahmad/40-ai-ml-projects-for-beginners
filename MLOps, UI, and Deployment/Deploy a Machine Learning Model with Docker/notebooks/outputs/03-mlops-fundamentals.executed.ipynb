**Definition:** MLOps is set of engineering practices to operationalize ML lifecycle end-to-end.

**Theory:** Explain mechanism, assumptions, and where this fits in deployment lifecycle.

**Motivation:** Why this matters for reliability, reproducibility, and operations.

**Real-World Example:** Team retrains monthly model, versions artifacts, deploys safely, and monitors drift.

**Visual Explanation:** See figure/code cell below.

**Code Explanation:** Code cell demonstrates concrete implementation details.

**Best Practices:** Track data/model/code/runtime as versioned assets.

**Common Mistakes:** Treating training notebook as final deliverable without deploy/monitoring plan.

```python
from pathlib import Path
import matplotlib.pyplot as plt

out = Path('outputs/figures')
out.mkdir(parents=True, exist_ok=True)

fig, ax = plt.subplots(figsize=(10, 3.5))
ax.axis('off')
stages = ['Develop', 'Train', 'Package', 'Deploy', 'Monitor', 'Maintain']
xs = [0.05, 0.2, 0.35, 0.5, 0.68, 0.84]
for x, stage in zip(xs, stages):
    ax.text(x, 0.6, stage, ha='center', va='center', bbox=dict(boxstyle='round', facecolor='#e6f4ea'))
for i in range(len(xs)-1):
    ax.annotate('', xy=(xs[i+1]-0.055, 0.6), xytext=(xs[i]+0.055,0.6), arrowprops=dict(arrowstyle='->'))
ax.text(0.5, 0.23, 'Docker strongest impact: Package + Deploy consistency', ha='center')
fig.tight_layout()
fig.savefig(out / 'mlops-lifecycle.png', dpi=150)
print('Saved:', out / 'mlops-lifecycle.png')
```
