## 1. Transformers and Self-Attention
Transformers replace recurrent computation with self-attention, letting every token attend to every other token in one layer.

For classification, encoder models output contextual token states, then classification head maps pooled representation to logits.

```python
from pathlib import Path
from domain_llm_ft.config.loader import load_config
cfg = load_config(Path('../configs/baseline.yaml'))
cfg
```

## 2. Tokenization
Tokenization maps text to IDs. Max length, truncation, and padding policy affect both accuracy and throughput.
Dynamic padding reduces wasted compute on short sequences.

```python
from domain_llm_ft.tokenization.factory import load_tokenizer
tok = load_tokenizer(cfg.tokenizer.name)
tok('Domain adaptation with PEFT and Transformers', truncation=True, max_length=32)
```
